```python
import numpy as np
def gradient_descent(x,y):
    m = b =0 
    # m is slope, b is intercept
    iterations = 1000
    n = len(x)
    learning_rate=0.001 # it depends your choice
    for i in range(iterations):
        y_d = m*x+b
        cost = (1/n)*sum([val**2 for val in(y-y_d)])
        md = -(2/n)*sum(x*(y-y_d))
        bd = -(2/n)*sum(y-y_d)
        m = m-learning_rate * md
        b = b-learning_rate * bd
        print("m {}, b {},cost {}, iteration {}".format(m,b,cost,i))    
x = np.array([1,2,3,4,5])
y = np.array([3,6,11,18,27]) # realtion is your choice
gradient_descent(x,y)
```
